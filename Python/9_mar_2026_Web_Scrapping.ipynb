```python
import requests
from bs4 import BeautifulSoup
import sys

def scrape_entire_html(url: str):
    """
    Fetches the HTML content of the target URL and prints the entire
    BeautifulSoup-formatted document to the console.

    Args:
        url: The web address to scrape.
    """
    print(f"--- Attempting to scrape HTML from: {url} ---")

    # Define a custom User-Agent to mimic a standard browser,
    # which can help prevent basic bot detection.
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }

    try:
        # 1. Fetch the page content using requests
        response = requests.get(url, headers=headers, timeout=15)

        # Raise an exception for bad status codes (4xx or 5xx)
        response.raise_for_status()

        # Ensure the content is decoded correctly
        content = response.content

        # 2. Parse the HTML content using BeautifulSoup
        # We use 'lxml' if available for speed, otherwise fallback to 'html.parser'
        try:
            soup = BeautifulSoup(content, 'lxml')
        except:
            soup = BeautifulSoup(content, 'html.parser')

        # 3. Print the entire, neatly formatted HTML document
        print("\n--- START OF ENTIRE HTML DOCUMENT (prettified) ---\n")
        print(soup.prettify())
        print("\n--- END OF ENTIRE HTML DOCUMENT ---\n")

    except requests.exceptions.HTTPError as e:
        print(f"\n🚨 HTTP Error (Client/Server issue): {e}", file=sys.stderr)
        print("This often means the server actively denied the request (e.g., 403 Forbidden).")
    except requests.exceptions.RequestException as e:
        print(f"\n🚨 Connection Error: {e}", file=sys.stderr)
        print("Check your network connection or if the website has strict IP/rate limiting.")
    except Exception as e:
        print(f"\n🚨 An unexpected error occurred: {e}", file=sys.stderr)

# --- Configuration ---
TARGET_URL = "https://www.amarpanchal.in/"

# --- Execution ---
if __name__ == "__main__":
    scrape_entire_html(TARGET_URL)
```
