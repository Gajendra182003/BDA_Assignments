üéØ Problem Statement

Create a function that calculates rental cost.

üìå Requirements:

Parameters:

Number of days

Cost per day

Default insurance charge per day = 100

Calculate total cost including insurance.

If rental days > 7, give 5% discount on total.

üí° Sample Call:
calculate_rent(5, 1000)
üì§ Expected Output:

Rent Cost: 5000

```python
def Car_Rental_Cost_Estimator(days,rental,insurance_charges=100):
    disscount=0
    if days>7:
        disscount=0.05
    total=days*rental+days*insurance_charges
    return total-(disscount*total)
```

Problem: ‚ÄúStartup Expense Tracker‚Äù
üéØ Problem Statement

Create a function that calculates total business expenses.

üìå Requirements:

Accept business name (mandatory parameter)

Accept any number of expenses using *kwargs

Print each expense category with amount

Print total expense

If no expenses are given, print "No expenses recorded."
office=15000
staff=150000
food=10000
tea=2000
ads=100000


```python
def startup_expens_tracker(name,*expenses):
    total_expense=0
    if len(expenses)==0:
        return f"{name}:No expenses recorded."
    else:
        print(f"Total expenses are:{len(expenses)}")
    for e in expenses:
        if e=="office":
            total_expense+=15000
        elif e=="staff":
            total_expense+=150000
        elif e=="food":
            total_expense+=10000
        elif e=="tea":
            total_expense+=2000
        elif e=="ads":
            total_expense+=100000
    return f"{name}:{total_expense}"
```

```python
startup_expens_tracker("CDAC","ads","office","tea")
```

```python
startup_expens_tracker("apna_startup")
```

```python
#Recursion :
#A way where a function calls itself.
#Comment on recursion:+ve and -ve both
#Recursion makes code look smaller, hence more readable.(+ve)
#It makes the program run slower, consumes more memory, and is very complex to understand at a time. (-ve)
```

```python
def factorial(no):
    if no==1:
        return 1
    else:
        return no*factorial(no-1)
```

```python
factorial(4)
```

```python
#Using recursion calculate GCD of two numbers.
#Following are the rules:
#If number one is perfectly divisible by number two, then number two is GCD.
#Otherwise let number one store number two and let number two store the remainder between number one and number two.
def GCD(no1,no2):
    if no1%no2==0:
        return no2
    else:
        return GCD(no2,no1%no2)
```

```python
GCD(18,16)
```

```python
#Global and local variables in Python.
#Variables declared in open space are global.
#Variables declared in some control structure are local.
#When a global and local variable have exactly the same identifier, preference is given to the local variable.
```

```python
#Global data can be accessed within the function.
x=100#global
def my_print():
    print(f"this is x:{x}")


print("main script")
my_print()
print("main script ends")
```

```python
#In order to allow accessing of global data for manipulation, we will use the keyword global.
x=100#global
def my_print():
    global x
    x=x+10
    print(f"\tInFunction this is x:{x}")


print("main script")
print(f"Main this is x:{x}")
my_print()
print(f"Main this is x:{x}")
print("main script ends")
```

```python
# prime number

def prime(num):
  if num <=1:
    return 1

  for i in range (2, num):
    if num %i==0:
      return ("Number is not prime !")
    else:
      return ("Number is prime !")


prime(27)
prime(21)
prime(80)
prime(29)
prime(13)
prime(15)
prime(31)
```
