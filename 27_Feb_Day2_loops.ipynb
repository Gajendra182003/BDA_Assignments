```python
#The infinite while loop.
#Loop eventually will not stop till the time the user wants to.
#Stop when the user gives nothing/blank
#When we stop, print:
'''
- the total number of even numbers
- the total number of odd numbers
- the sum of even numbers
- the sum of odd numbers
- the grand total, adding both sums'''
sum_even=sum_odd=count_even=count_odd=0
while True:
    data=input("Enter data:")
    if data=="":
        break#Break statement stops execution of loop.
    else:
        data=float(data)
        if data%2==0:#even
            count_even+=1
            sum_even+=data
        else:
            count_odd+=1
            sum_odd+=data

print(f"Total number of even entries {count_even}\nTotal number of odd entries {count_odd}\nAddition of even numbers {sum_even}\nAddition of odd numbers {sum_odd}\nGrand total {sum_even+sum_odd}")
```

<h2>üìà Compound Interest Formula (Per Year) for n Years</h2>

<h3>üìå Formula</h3>
<p>
A = P √ó (1 + R/100)<sup>n</sup>
</p>

<h3>üìñ Where:</h3>
<ul>
<li><b>A</b> = Final Amount after n years</li>
<li><b>P</b> = Principal (Initial Investment)</li>
<li><b>R</b> = Annual Interest Rate (in %)</li>
<li><b>n</b> = Number of Years</li>
</ul>

<h3>üßÆ Step-by-Step Understanding</h3>
<p>
After 1 year:<br>
P √ó (1 + R/100)
</p>

<p>
After 2 years:<br>
P √ó (1 + R/100)<sup>2</sup>
</p>

<p>
After n years:<br>
P √ó (1 + R/100)<sup>n</sup>
</p>

<h3>üí° Interest Earned Only</h3>
<p>
Compound Interest (CI) = A ‚àí P
</p>

<h3>üìå Example</h3>
<p>
Principal = ‚Çπ50,000 <br>
Rate = 8% <br>
Time = 10 years
</p>

<p>
A = 50000 √ó (1.08)<sup>10</sup> <br>
A ‚âà ‚Çπ1,07,946
</p>

```python
principal = float(input("Enter Principal Amount: "))
rate = float(input("Enter Annual Interest Rate (%): "))
years = int(input("Enter Number of Years: "))

balance = principal

print("\nYear-wise Growth:")
for year in range(1, years + 1):
    balance = balance * (1 + rate / 100)
    print("Year", year, "Balance:", balance)
    if balance >= 100000:
        print("\nTarget of 100000 reached in", year, "years!")
        break
```

```python
monthly_budget=float(input("Enter your monthly budget:"))
warning_amount=monthly_budget*0.8
total_spent_amount=0
remaining_balance=monthly_budget
while True:
    print("\nSmart Expense & Budget Tracker")
    print("--------------------------------")
    print("1. Add Expense")
    print("2. View Total Spent")
    print("3. View Remaining Budget")
    print("4. Exit")
    print("--------------------------------")
    choice=int(input(":"))
    if choice==1:
        expense=float(input("Enter Expense Amount:"))
        if remaining_balance>=expense:
            total_spent_amount+=expense
            remaining_balance-=expense
            print(f"\nExpense Updated. New Balance left:{remaining_balance}")
            if total_spent_amount>=warning_amount:
                print("‚ö† Warning: Budget Almost Full (80% used)")
        else:
            print(f"\n Expense cannot be added. Reason: remaining balance:{remaining_balance}")
    elif choice==2:
        print(f"\nTotal amount spent till now {total_spent_amount} from total {monthly_budget} left is {monthly_budget-total_spent_amount}")
    elif choice==3:
        print(f"\nTotal amount of {monthly_budget} left is {monthly_budget-total_spent_amount}")
    elif choice==4:
        break
    else:#Default else if for error trapping. Simply informs the user that wrong input was given.
        print("\n please check the option selected.")
```
