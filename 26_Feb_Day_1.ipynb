```python
#Assignment takes place from R value to L value.
x=10
y=5+5
z=13-3
#id(): this command provides symbolic reference/logical address or reference to the variable.
print(f"x={x} @{id(x)}\t y={y} @{id(y)}\t z={z} @{id(z)}")
#In order to manage memory more effectively, similar data for different identifiers shares same address space.
x=x+1
print(f"x={x} @{id(x)}\t y={y} @{id(y)}\t z={z} @{id(z)}")
y=y*2
print(f"x={x} @{id(x)}\t y={y} @{id(y)}\t z={z} @{id(z)}")
z=z-5
print(f"x={x} @{id(x)}\t y={y} @{id(y)}\t z={z} @{id(z)}")
```

```python
Major use of if-else-if is in the range-based system.
#Reading the percentage print result:
- 60 or greater: first class
- Less than 60 but greater than or equal to 50: second class
- Less than 50 but greater than or equal to 40: third class
- Less than 40: fail
if
elif
elif
else
```

```python
# Reading percentage from user
percentage = float(input("Enter your percentage: "))#0
#In an if-else-if ladder at most one block only gets executed.
# Range-based result system
if percentage >= 60:
    print("Result: First Class")

elif percentage >= 50 and percentage < 60:
    print("Result: Second Class")

elif percentage >= 40 and percentage < 50:
    print("Result: Third Class")

else:
    print("Result: Fail")
```
