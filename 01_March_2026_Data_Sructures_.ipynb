```python
What is Deep Copy (making an actual copy of data which is independent )
and Shallow Copy(giving alternate ID to the list) in a list?
```

```python
#Shallow
data=[11,22,33,44]
d1=data
print(f"data:{data}@{id(data)}  d1:{d1}@{id(d1)}")
d1[0]=d1[0]*100
print(f"data:{data}@{id(data)}  d1:{d1}@{id(d1)}")
```

```python
#Deep
data=[11,22,33,44]
d1=data.copy()
print(f"data:{data}@{id(data)}  d1:{d1}@{id(d1)}")
d1[0]=d1[0]*100
print(f"data:{data}@{id(data)}  d1:{d1}@{id(d1)}")
```

```python
def smallest_n(data,n):
    return sorted(data)[n-1]
def largest_n(data,n):
    return sorted(data)[::-1][n-1]
```

```python
n=int(input("Enter n:"))
data=[88,11,33,22,77,55,8,24,67,98,43]
print(f"{n} smallest element is {smallest_n(data,n)}")
print(f"{n} largest element is {largest_n(data,n)}")
print(f"{sorted(data)}")

```

```python
#Shopping list application
Insert an element, delete an element, print in list.
While inserting check if it does not exist; then only insert.
While deleting check if it exists then only delete.
Print is a normal printing listing all the elements entered in the shopping list.
```

```python
def menu():
    print("\n-----------------------")
    print("   My Shopping List App")
    print("-----------------------")
    print("1.Enter element in List")
    print("2.Delete element in List")
    print("3.Print The List")
    print("4.Exit")
    print("-----------------------")

def add_item(item,shopping_list):
    if item not in shopping_list:
        shopping_list.append(item)
        print(f"{item} added")
    else:
        print(f"{item} already there")
def print_list(shopping_list):
    print("\nItems in list are")
    print("-------------------")
    for index in range(len(shopping_list)):
        print(f"{index+1}.{shopping_list[index]}")

def delete_list(shopping_list):
    print_list(shopping_list)
    item_number=int(input("Enter item number to delete:"))
    if 0<item_number<=len(shopping_list):
        print(f"{shopping_list.pop(item_number-1)} deleted")
    else:
        print(f"{item_number} seems to be out of range.")
```

```python
shopping_list=[]
choice=0
while True:
    menu()#We will print all the options of the menu.
    choice=int(input(":"))
    if choice==1:
        item=input("Enter Item to add:")
        add_item(item,shopping_list)
    elif choice==2:
        delete_list(shopping_list)
    elif choice==3:
        print_list(shopping_list)
    elif choice==4:
        print("Thanks for using the system.")
        break
    else:
        print("Invalid option.")


```

```python
#Perform clockwise rotation of a content in a list by n elements where n is given by the user.
#sample [10,20,30,40,50]--->2 rotations [20,30,40,50,10]-->[30,40,50,10,20]
#Give me a method which can remove the first element from the list. pop(0)
#Give me a method which will insert an element at the end of the list. :append()
```

```python
def rotate_list(data,n):
    for i in range(n):#If start is not given: zero /if step is not given +1.
        temp=data.pop(0)#Removing the first element
        data.append(temp)#Adding data to the end of the list
        print(data)#Print list.

data=[10,20,30,40,50]
n=int(input("Enter number of time rotations needed:"))
rotate_list(data,n)
```
